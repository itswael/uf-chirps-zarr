# Shapefile Analysis

This notebook explores the content of shape files in the sample_data directory.
Raw shapefile structure and metadata will be examined.

```python
import geopandas as gpd
import os
import sys

# Set up paths
sample_data_path = '../frontend/sample_data'
```

```python
# List available shape files
print("Available files in sample_data directory:")
for file in os.listdir(sample_data_path):
    file_path = os.path.join(sample_data_path, file)
    file_size = os.path.getsize(file_path)
    print(f"  {file} ({file_size} bytes)")
```

## Sri Lanka Shapefile Analysis

```python
# Load Sri Lanka shapefile
sri_lanka_path = os.path.join(sample_data_path, 'Sri_Lanka.shp')
print(f"Loading: {sri_lanka_path}")
sri_lanka_gdf = gpd.read_file(sri_lanka_path)

print("\n=== Sri Lanka Shapefile Content ===")
print(f"Shape: {sri_lanka_gdf.shape}")
print(f"\nColumns: {sri_lanka_gdf.columns.tolist()}")
print(f"\nData types:\n{sri_lanka_gdf.dtypes}")
print(f"\nCRS: {sri_lanka_gdf.crs}")
```

```python
# Display raw data
print("\n=== Raw Data ===")
print(sri_lanka_gdf)
```

```python
# Geometry details
print("\n=== Geometry Details ===")
for idx, row in sri_lanka_gdf.iterrows():
    print(f"\nRow {idx}:")
    print(f"  Geometry type: {row.geometry.geom_type}")
    print(f"  Bounds: {row.geometry.bounds}")
    print(f"  Area: {row.geometry.area}")
    print(f"  Length: {row.geometry.length}")
```
